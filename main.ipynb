```python
import pandas as pd

last_month_billing = pd.read_csv('last_month_billing.csv')
current_customers = pd.read_csv('current_customers.csv')
all_time_billing = pd.read_csv('all_time_billing.csv')
```

```python
all_time_combined = pd.merge(all_time_billing, current_customers, on='Company ID', how='inner')
last_month_combined = pd.merge(last_month_billing, current_customers, on='Company ID', how='inner')
```

```python
all_time_combined.head()
```

```python
last_month_combined.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
target = 'Total Billing Across Portfolio Last Month'
features = last_month_combined.select_dtypes(include=[np.number]).columns.tolist()
features.remove(target)

numeric_columns = last_month_combined.select_dtypes(include=[np.number]).columns
last_month_combined[numeric_columns] = last_month_combined[numeric_columns].fillna(last_month_combined[numeric_columns].mean())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    last_month_combined[features],
    last_month_combined[target],
    test_size=0.2,
    random_state=42
)
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(random_state=42)
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results[name] = {"MSE": mse, "R^2": r2}

results

```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
TreeReg = DecisionTreeRegressor()
TreeReg.fit(X_train, y_train)

y_pred_Train = TreeReg.predict(X_train) #Predictions on training set
y_pred_Test = TreeReg.predict(X_test)   #Predictions on testing set
```

## Train Metrics

```python
from sklearn import metrics
print('R squared:', metrics.r2_score(y_train, y_pred_Train))
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_train, y_pred_Train))  
```

```python
print('Mean Squared Error:', metrics.mean_squared_error(y_train, y_pred_Train))  
```

```python
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_train, y_pred_Train)))
```

## Test Metrics

```python
print('R squared:', metrics.r2_score(y_test, y_pred_Test))
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred_Test))  
```

```python
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred_Test))  
```

```python
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_Test)))
```

# KNN (K-Nearest Neighbors)

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

knn_model = KNeighborsRegressor()
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)
mse_knn = mean_squared_error(y_test, y_pred_knn)
r2_knn = r2_score(y_test, y_pred_knn)

print("KNN Model:")
print(f"MSE: {mse_knn}")
print(f"R^2: {r2_knn}")
```

# Random Forest:

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("Random Forest Model:")
print(f"MSE: {mse_rf}")
print(f"R^2: {r2_rf}")
```

# Extra Trees:

```python
from sklearn.ensemble import ExtraTreesRegressor

et_model = ExtraTreesRegressor(random_state=42)
et_model.fit(X_train, y_train)
y_pred_et = et_model.predict(X_test)
mse_et = mean_squared_error(y_test, y_pred_et)
r2_et = r2_score(y_test, y_pred_et)

print("Extra Trees Model:")
print(f"MSE: {mse_et}")
print(f"R^2: {r2_et}")
```

# Gradient Boosting Regressor (GBR):

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr_model = GradientBoostingRegressor(random_state=42)
gbr_model.fit(X_train, y_train)
y_pred_gbr = gbr_model.predict(X_test)
mse_gbr = mean_squared_error(y_test, y_pred_gbr)
r2_gbr = r2_score(y_test, y_pred_gbr)

print("Gradient Boosting Regressor Model:")
print(f"MSE: {mse_gbr}")
print(f"R^2: {r2_gbr}")
```

# XGBoost Regressor (XGBR):

```python
from xgboost import XGBRegressor

xgbr_model = XGBRegressor(random_state=42)
xgbr_model.fit(X_train, y_train)
y_pred_xgbr = xgbr_model.predict(X_test)
mse_xgbr = mean_squared_error(y_test, y_pred_xgbr)
r2_xgbr = r2_score(y_test, y_pred_xgbr)

print("XGBoost Regressor Model:")
print(f"MSE: {mse_xgbr}")
print(f"R^2: {r2_xgbr}")
```

# CatBoost

```python
from catboost import CatBoostRegressor

catboost_model = CatBoostRegressor(random_state=42, verbose=0)
catboost_model.fit(X_train, y_train)
y_pred_catboost = catboost_model.predict(X_test)
mse_catboost = mean_squared_error(y_test, y_pred_catboost)
r2_catboost = r2_score(y_test, y_pred_catboost)

print("CatBoost Model:")
print(f"MSE: {mse_catboost}")
print(f"R^2: {r2_catboost}")
```

# LightGBM:

```python
from lightgbm import LGBMRegressor

lgbm_model = LGBMRegressor(random_state=42)
lgbm_model.fit(X_train, y_train)
y_pred_lgbm = lgbm_model.predict(X_test)
mse_lgbm = mean_squared_error(y_test, y_pred_lgbm)
r2_lgbm = r2_score(y_test, y_pred_lgbm)

print("LightGBM Model:")
print(f"MSE: {mse_lgbm}")
print(f"R^2: {r2_lgbm}")
```
